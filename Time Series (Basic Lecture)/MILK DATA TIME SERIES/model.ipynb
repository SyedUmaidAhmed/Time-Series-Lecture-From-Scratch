```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

from datetime import datetime

```

```python
data=pd.read_csv('monthly-milk-production-pounds.csv')
data.shape
```

```python
data.head()
```

```python
data.tail()
```

```python
data.drop(169,axis=0,inplace=True)
```

```python
data.tail()
```

```python
data['Month']=pd.to_datetime(data['Month'])


```

```python
data=data.set_index(['Month']) #Make it as Index
```

```python
data.head()
```

```python
data.tail()
```

```python
data.index
```

```python
data.drop(index=data.index[-1], axis=0, inplace=True)
```

```python
data.tail()
```

```python
data.plot()
```

```python
rolmean=data.rolling(window=12).mean()
rolstd=data.rolling(window=12).std()
```

```python
plt.plot(data, color='red', label='Actual')

plt.plot(rolmean, color='green', label='Rolling Mean') 

plt.plot(rolstd, color='black', label='Rolling Std')

plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```
